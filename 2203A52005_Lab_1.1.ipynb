```python
import pandas as pd
import numpy as np

# Load dataset
insurance_data = pd.read_csv("insurance.csv")

#vector
ages=insurance_data['age'].tolist()

#factor
smoking_status=pd.Categorical(insurance_data['smoker'])

example_list=[ages, smoking_status]

#matrix
expenses_matrix=np.array(insurance_data['expenses']).reshape(-1, 1)

#DataFrame
df_example = pd.DataFrame({
    'Ages': ages,
    'Smoking': smoking_status,
    'Expenses': insurance_data['expenses']
})

print("Vector (ages):", ages[:5])
print("Factor (smoking_status):", smoking_status[:5])
print("List:", example_list[:2])
print("Matrix (expenses):\n", expenses_matrix[:2])
print("DataFrame:\n", df_example.head())
```

```python
#Plot a scatter plot between age and smoking
import matplotlib.pyplot as plt

plt.scatter(insurance_data['age'], insurance_data['smoker'].map({'yes': 1, 'no': 0}))
plt.title("Scatter Plot of Age vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoking Status (1=Yes, 0=No)")
plt.grid(True)
plt.show()

```

```python
# Filter data for age 20-40
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]
filtered_smoker_numeric = filtered_data['smoker'].map({'yes': 1, 'no': 0})

# Scatter plot: Age (20-40) vs Smoking
plt.scatter(filtered_data['age'], filtered_smoker_numeric, c='green', marker='x')
plt.title("Scatter Plot of Age (20-40) vs Smoking")
plt.xlabel("Age (20-40)")
plt.ylabel("Smoking Status (1=Yes, 0=No)")
plt.grid(True)
plt.show()
```

```python
# Calculate mean expenses for smokers and non-smokers
expenses_by_smoking = insurance_data.groupby('smoker')['expenses'].mean()

# Bar plot for mean expenses
expenses_by_smoking.plot(kind='bar', color=['orange', 'blue'])
plt.title("Mean Expenses: Smokers vs Non-Smokers")
plt.xlabel("Smoking Status")
plt.ylabel("Mean Expenses")
plt.show()
```

```python
# Count number of smokers by gender
smokers_gender = insurance_data[insurance_data['smoker'] == 'yes'].groupby('sex').size()

# Bar plot for smokers gender-wise
smokers_gender.plot(kind='bar', color=['purple', 'green'])
plt.title("Number of Smokers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Smokers")
plt.show()
```

```python
# Scatter plot: Age vs Expenses
plt.scatter(insurance_data['age'], insurance_data['expenses'], c='purple', alpha=0.5)
plt.title("Scatter Plot of Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.grid(True)
plt.show()
```
